Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

Закодируем категориальные признаки  с помощью one-hot-encoding'а.



```python
ohe = OneHotEncoder(drop='if_binary', sparse_output=False)
ohe.fit(data_train[cat_cols])

# названия новых колонок после one-hot-encoding
ohe_feature_names = ohe.get_feature_names_out(cat_cols)

# преобразуем  numpy массив в DataFrame
new_category_cols_df = pd.DataFrame(ohe.transform(data_train[cat_cols]), columns=ohe_feature_names, index=data_train.index)

train_data = data_train.drop(columns=cat_cols)
train_data = pd.concat([train_data, new_category_cols_df], axis=1)
```

```python
for col in train_data.columns:
  dif_val = train_data[col].unique()
  print(f'{col} = {dif_val}')
```

```python
correlation_matrix = train_data.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.show()
```

```python
X = train_data.drop(columns='Churn')
y = train_data[target_col]


pipeline = make_pipeline(
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)

param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}

grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5, # 5-кратная кроссвалидация
    scoring='roc_auc', # метрика
    n_jobs=-1, # все доступные ядра процессора
    verbose=1
)

grid_search.fit(X, y)

print(f"Лучшие параметры: {grid_search.best_params_}")
print(f"Лучший ROC-AUC на кроссвалидации: {grid_search.best_score_:.4f}")

best_model = grid_search.best_estimator_
```

```python
from catboost import CatBoostClassifier, Pool

x = data_train.drop(columns=target_col)
y = data_train[target_col]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
params= {
    'iterations': [100, 200, 300, 400, 500],
    'learning_rate': [0.01, 0.1]
}

best_score = 0
best_params = None
best_model = None
```

```python
for it in params['iterations']:
  for lr in params['learning_rate']:
    model = CatBoostClassifier(
        iterations=it,
        learning_rate=lr,
        verbose=False,
        random_seed=42
    )
    model.fit(x_train, y_train, cat_features=cat_cols)
    pred = model.predict_proba(x_test)[:, 1]
    score = roc_auc_score(y_test, pred)

    print(f'iterations = {it}, learning_rate = {lr}, score = {score}')

    if score > best_score:
      best_score = score
      best_params = (it, lr)
      best_model = model
```

Смотрим на конечный результат

```python
print('Best ROC-AUC:', best_score)
print('Best params:', best_params)
print('Best_model:', best_model)
```
